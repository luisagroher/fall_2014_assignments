```python
cases = []
deaths = []
df.set_index = ['date']
```

```python
for column in df.columns:
    tmp = column.split('_')
    if tmp[0]  == 'Cases':
        cases.append(column)
    if tmp[0] == 'Deaths':
        deaths.append(column)
```

```python
cases_data = df[cases]
deaths_data = df[deaths]
```

```python
cases_data.columns = ['Guinea', 'Liberia', 'SierraLeone', 'Nigeria', 'Senegal', 'UnitedStates', 'Spain', 'Mali']
deaths_data.columns = ['Guinea', 'Liberia', 'SierraLeone', 'Nigeria', 'Senegal', 'UnitedStates', 'Spain', 'Mali']
```

```python
cases_data['group'] = cases_data.apply(lambda col: 'cases')
cases_data['group'] = cases_data['group'].fillna('cases')
deaths_data['group'] = deaths_data.apply(lambda col: 'deaths')
deaths_data['group'] = deaths_data['group'].fillna('deaths')
```

```python
data = cases_data
data = data.append(deaths_data)
```

```python
data.index = pd.to_datetime(data.index)
```

```python
#### TimeSeriesModeller from lecture


from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error


class TimeSeriesModeller:
    def __init__(self):
        self.model = LinearRegression()
        self.results = {}
        
    def fit(self,y):
        for ar_size in xrange(1, 10):
            train_start = int(len(y)/2)
            lagged = pd.concat([y.shift(i) for i in range(1,ar_size+1)], axis=1)
            self.model.fit(lagged.ix[11:train_start],y.ix[11:train_start])
            predictions = []
            for t in range(train_start, len(y)):
                prediction = self.model.predict(lagged.ix[t])
                predictions.append(abs(prediction/y.ix[t]-1))
            error = mean(predictions)
            #errors = []
            #errors.append(errors)
            self.results[ar_size] = {
                        'model size': ar_size,
                       'mse': error}
        
        return self.results
            #print "model size %f" %(ar_size)      
            #print "meansquarederror %f"%(error)
```
